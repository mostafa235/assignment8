```python
import requests
from bs4 import BeautifulSoup
import csv

```

```python
url="https://books.toscrape.com/index.html"
soup=BeautifulSoup(requests.get(url).text,'html.parser')

rate1={"One":1,"Two":2,"Three":3,"Four":4,"Five":5}

with open ("books.csv","w",newline="") as f:

    writer=csv.writer(f)
    writer.writerow(["BookName","rate","salary","category"])
    for i in soup.find_all("article",attrs={"class":"product_pod"}):
        name=i.find("h3").find("a").get("title")
        rate=rate1[i.find("p").get("class")[1]]
        salary=i.find("p",attrs={"class":"price_color"}).get_text()[2:]
        category=soup.find("title").get_text().split("|")[0].strip()
        writer.writerow([name,rate,salary,category])
```

```python
# save data of category without next pages
url="https://books.toscrape.com/index.html"
soup=BeautifulSoup(requests.get(url).text,'html.parser')

#
rate1={"One":1,"Two":2,"Three":3,"Four":4,"Five":5}
#open csv fiel
with open ("books.csv","w",newline="",encoding="utf-8") as f:
    
            
    writer=csv.writer(f)
    writer.writerow(["BookName","rate","salary","category"])
    # category loop
    for i1 in soup.find("ul",attrs={"class":""}).find_all("li"):
        
        url_domain=url[:url.rfind("/")+1]

        category=url_domain+i1.find("a").get("href")
        urll=category
        
        
            
            
        soup=BeautifulSoup(requests.get(urll).text,'html.parser')
        
        
        for i3 in soup.find_all("article",attrs={"class":"product_pod"}):
    
                
            name=i3.find("h3").find("a").get("title")
            
                
                
            rate=rate1[i3.find("p").get("class")[1]]
                
            salary=i3.find("p",attrs={"class":"price_color"}).get_text()[2:]
                
            category1=soup.find("title").get_text().split("|")[0].strip()
            writer.writerow([name,rate,salary,category1])
                
```

```python
# save data of category with next pages
url="https://books.toscrape.com/index.html"
soup=BeautifulSoup(requests.get(url).text,'html.parser')

#
rate1={"One":1,"Two":2,"Three":3,"Four":4,"Five":5}
#open csv fiel
with open ("books.csv","w",newline="",encoding="utf-8") as f:
    
            
    writer=csv.writer(f)
    writer.writerow(["BookName","rate","salary","category"])
    # category loop
    for i1 in soup.find("ul",attrs={"class":""}).find_all("li"):
        
        url_domain=url[:url.rfind("/")+1]

        category=url_domain+i1.find("a").get("href")
        urll=category
        
        
            
            
        soup=BeautifulSoup(requests.get(urll).text,'html.parser')
        
        
        for i3 in soup.find_all("article",attrs={"class":"product_pod"}):
    
                
            name=i3.find("h3").find("a").get("title")
            
                
                
            rate=rate1[i3.find("p").get("class")[1]]
                
            salary=i3.find("p",attrs={"class":"price_color"}).get_text()[2:]
                
            category1=soup.find("title").get_text().split("|")[0].strip()
            writer.writerow([name,rate,salary,category1])
            while soup.find("li",attrs={"class":"next"}):    
                page_next=urll[:urll.rfind("/")+1]+soup.find("li",attrs={"class":"next"}).find("a").get("href")
                soup=BeautifulSoup(requests.get(page_next).text,'html.parser')
                                 
        
            #loop for page  save 
                for i3 in soup.find_all("article",attrs={"class":"product_pod"}):
                                 
                
                    name=i3.find("h3").find("a").get("title")
                
                    rate=rate1[i3.find("p").get("class")[1]]
                
                    salary=i3.find("p",attrs={"class":"price_color"}).get_text()[2:]
                
                    category1=soup.find("title").get_text().split("|")[0].strip()
                
                    writer.writerow([name,rate,salary,category1])
                
```
